```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
import folium
```

```python
df = pd.read_csv('../datasets/imoveis/copacabana_lat_lng.csv')
```

```python
df.head()
```

```python
plt.scatter(df['lng'], df['lat'], marker='.')
```

```python
CENTRO = (-22.9732508, -43.1945315)
mapa = folium.Map(CENTRO, zoom_start=14)
for i in range(200):
    linha = df.iloc[i]
    if linha['lat'] and linha['lng']:
        folium.Marker(
            (linha['lat'], linha['lng']), popup='Quartos: {}'.format(linha['Quartos'])
        ).add_to(mapa)
display(mapa)
```
