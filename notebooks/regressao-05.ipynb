```python
def limpar_valor(valor):
    try:
        valor = float(valor.replace("R$", "").replace(".", "").replace(",", ".").strip())
    except AttributeError:
        valor = np.nan
    return valor
```

```python
dados["preço"].apply(limpar_valor).tail()
```

```python
dados["preço"] = dados["preço"].apply(limpar_valor)
dados["condomínio"] = dados["condomínio"].apply(limpar_valor)
dados["iptu"] = dados["iptu"].apply(limpar_valor)
```

```python
dados["condomínio"].isnull() == False
```

```python
dados = dados[(dados["condomínio"].isnull() == False) & (dados["quartos"].isin([2, 3]))]
```

```python
len(dados)
```

```python
dados_completos = dados[dados["iptu"].isnull() == False]
```

```python
len(dados_completos)
```

```python
media_iptu = dados_completos["iptu"].mean()
media_area = dados_completos["área"].mean()
```

```python
media_area
```

```python
media_iptu
```

```python
media_iptu_area = media_iptu / media_area
```

```python
media_iptu_area
```

```python
def calcula_iptu(linha):
    if pd.isnull(linha["iptu"]):
        return media_iptu_area * linha["área"]
    return linha["iptu"]
```

```python
dados["iptu"] = dados.apply(calcula_iptu, axis=1)
```
